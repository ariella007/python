1) create first array "overs" with 4 elements
5,10,15,20
create another array "runs_scored" with 4 elements and store
no.of runs scored at the end of each over
e.g.  40 runs at the end of 5th over , 75 runs at the end of 10th over etc.

now create a line plot to show the relationship between "overs" and "runs_scored"

```python
from turtle import color
from numpy import *
from matplotlib import pyplot as py
over=array([5,10,15,20])
runs_scored=array([40,75,135,180])
py.plot(over,runs_scored,color='m',linestyle=':',linewidth=1)
py.title("the relationship between overs and runs_scored")
py.xlabel("overs")
py.ylabel("scores")
py.grid(True)
py.show()
```

2) create a Bar plot to show how many people like "Action","Romance","Comedy" or "Drama" movies.

```python
from matplotlib import pyplot as py
mylist={"Action":70,"Romance":45,"Comedy":30,"Drama":80}
names=list(mylist.keys())
marks=list(mylist.values())
py.bar(names,marks)
py.xlabel("Intrest")
py.ylabel("rate of people")
py.show()
```

3) create a piechart to show popularity of various modules (Java,Python etc.)

```python
from matplotlib import pyplot as py
lang=['java','Python','os','Html','jscript']
popularity=[35,25,5,25,10]
py.pie(popularity,labels=lang,autopct='%0.1f%%',colors=['red','green','pink','blue','yellow'])
py.legend(bbox_to_anchor=(1.3,0.1))
py.show()
```
